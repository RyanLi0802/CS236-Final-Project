```python
from PIL import Image

def gif_to_frames(gif_path, output_folder):
    # Open the gif
    img = Image.open(gif_path)
    print(img.n_frames)
    # Loop over each frame in the gif
    for frame in range(0, img.n_frames):
        # Seek to the frame in the gif
        img.seek(frame)

        # Save the frame as an image file
        img.save(f'{output_folder}/frame_{frame}.png')

# Usage
gif_to_frames(waterfall, output_dir)
```

```python
from PIL import Image, ImageDraw, ImageFont


def concatenate_images(image_paths, output_path, ):
    images = [crop_image(x, 150) for x in image_paths]

    widths, heights = zip(*(i.size for i in images))

    total_width = sum(widths)
    max_height = max(heights)

    new_image = Image.new('RGB', (total_width, max_height))

    x_offset = 0
    for img in images:
        new_image.paste(img, (x_offset, 0))
        x_offset += img.width

    new_image.save(output_path)


def concatenate_images_captions(image_paths, output_path, caption):
    images = [Image.open(x) for x in image_paths]

    widths, heights = zip(*(i.size for i in images))

    total_width = sum(widths)
    max_height = max(heights)

    # Adjust the height of the new image to include space for the caption
    new_image = Image.new('RGB', (total_width, max_height + 50))  # adjust the "+ 50" as needed for caption space

    x_offset = 0
    for img in images:
        new_image.paste(img, (x_offset, 50))  # adjust the "50" to match the space you've added for the caption
        x_offset += img.width

    draw = ImageDraw.Draw(new_image)

    # Use a truetype font
    font = ImageFont.truetype('/Library/Fonts/Arial.ttf', 15)  # adjust the font and size as needed

    text_width, text_height = draw.textsize(caption, font)
    text_x = (total_width - text_width) / 2
    text_y = (50 - text_height) / 2  # adjust the "50" to match the space you've added for the caption

    draw.text((text_x, text_y), caption, font=font, fill="white")

    new_image.save(output_path)


def crop_image(image_path, buffer=100):
    img = Image.open(image_path)
    width, height = img.size
    print(width, height)

    # If the image is already a square, there's nothing to do
    if width == height:
        return img

    # If the image is wider than it is tall, crop the width
    if width > height:
        left = buffer
        top = 0
        right = height + buffer
        bottom = height
        img_cropped = img.crop((left, top, right, bottom))
        return img_cropped
    else:
        print("Image is not wider than its height. No cropping performed.")
        return img


def gif(name):
    # image_path = f"/content/drive/Shareddrives/CS236/final_project/frames/{name}/frame_"
    image_path = f"/content/drive/Shareddrives/CS236/final_project/main/results_fewshot/frames/{name}/"
    # image_path = f"/content/drive/Shareddrives/CS236/final_project/frames/{name}/"
    # build image paths
    image_paths = []
    for i in range(8):

        image_paths.append(f"{image_path}frame_{int(i * 2)}.png")
    output_path = f'/content/drive/Shareddrives/CS236/final_project/main/results_fewshot/frames/final_report/{name}.jpg'
    concatenate_images(image_paths, output_path)


def video(name):
    image_path = f"/content/drive/Shareddrives/CS236/final_project/main/DAVIS/JPEGImages/480p/rhino/"
    # build image paths
    image_paths = []
    for i in range(4):
        image_paths.append(f"{image_path}0000{int(i)}.jpg")
    output_path = f'/content/drive/Shareddrives/CS236/final_project/concatenated/{name}/{name}.jpg'
    concatenate_images(image_paths, output_path)


# Usage
if __name__ == '__main__':
    # gif("boat_watercolor")
    # gif("car_disney")
    # gif("car_desert")
    # gif("yacht")
    # gif("no_inv")
    #video("rhino_2")
    #gif("lion_1")
    gif("waterfall_2")
```
